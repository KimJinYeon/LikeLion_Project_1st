```python
day_reported = pd.read_csv('./data/data_2021-Sep-28_day_reported.csv')
test_conducted = pd.read_csv('./data/data_2021-Sep-28_test_conducted.csv')
death_with_certificate = pd.read_csv('./data/data_2021-Sep-28_death_with_certificate_daily.csv')
patients_admitted = pd.read_csv('./data/data_2021-Sep-28_patients_admitted.csv')
patients_ICU = pd.read_csv('./data/data_2021-Sep-28_patients_ICU.csv')
patients_in_hospital = pd.read_csv('./data/data_2021-Sep-28_patients_in_hospital.csv')
vaccin_first = pd.read_csv('./data/data_2021-Sep-28_vaccin_first.csv')
vaccin_second = pd.read_csv('./data/data_2021-Sep-28_vaccin_second.csv')
vaccin_uptake = pd.read_csv('./data/data_2021-Sep-28_vaccin_uptake.csv')
```
